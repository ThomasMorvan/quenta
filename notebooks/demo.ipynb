```
from quenta import get_settings, io, plots
from quenta.analysis import fit_per_session, summarize_sessions

settings = get_settings()
paths = settings.paths.ensure()
print('data_root :', paths.root)
print('resolved from:', ', '.join(settings.sources))
```

## 1. Data with `io/` module

```
# load data, etc.
io.make_fake_dataset(paths)
raw = io.load_raw(paths)
print(raw.shape, '|', raw['session'].nunique(), 'sessions')
raw.head()
```

## 2. Analysis with `analysis/` module

```
# data in, numbers out
fits = fit_per_session(raw)
summary = summarize_sessions(fits)

print(f"slope {summary['slope_mean']:.3f} "
      f"[{summary['slope_ci_lo']:.3f}, {summary['slope_ci_hi']:.3f}] "
      f"| median r2 {summary['r2_median']:.3f}")
fits.head()
```

```
(fits[['session', 'slope', 'true_slope', 'intercept', 'true_intercept', 'r2']]
 .round(3))
```

## 3. Figures with `plots/` module

```
# data in, figure out
fig = plots.figure_overview(raw, fits, summary)
fig
```

```
written = plots.save_overview(fig, paths)
for p in written:
    print(p)
```
